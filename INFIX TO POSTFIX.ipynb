```python
#INFIX TO POSTFIX

def infixtopostfix(exp):
    p={"+":1,"-":1,"*":2,"/":2,"^":3,"(":0}
    fin=""
    op=[]
    for i in exp:
        if i == "(":
            op.append(i)
        elif i==")":
            while(op[-1]!=")"):
                fin+=op.pop()
            op.pop()
        elif i in p:
            while len(op)>0 and p[op[-1]]>=p[i]:
                fin+=op.pop()
            op.append(i)
        else:
            fin+=i
    while len(op)>0:
        fin+=op.pop()
    return fin
print(infixtopostfix("A+B*C+D"))
```
